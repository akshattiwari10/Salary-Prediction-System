##Experiment v2 â€“ City-Based Feature (Prototype)
City values are randomly assigned to validate pipeline scalability.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/drive/MyDrive/Salary Predictor/Salary Data.csv")
print(df.head())
```

```python
cities= ['Banglore', 'Mumbai','Delhi','Hyerabad','Pune','Other']
df['city']=np.random.choice(cities, size=len(df))
```

```python
df[['city']].head()
```

```python
df[['Age', 'Years of Experience', 'Salary']].head(10)
```

```python
df.info()
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Years of Experience'] = df['Years of Experience'].fillna(df['Years of Experience'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
```

```python
df.info()
```

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
df['Education Level']=df['Education Level'].fillna(df['Education Level'].mode()[0])
df['Job Title']=df['Job Title'].fillna(df['Job Title'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#seperating features from target
X= df.drop('Salary', axis=1)  #features
y=df['Salary']                # salary
```

```python
categorical_cols = ['Gender', 'Education Level', 'Job Title','city']
numerical_cols=['Age', 'Years of Experience']
```

```python
#applying one hot encoding
x_encoded=pd.get_dummies(X, columns=categorical_cols, drop_first=True)
```

```python
x_encoded.head()
```

```python
#standard Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_encoded[numerical_cols]=scaler.fit_transform(x_encoded[numerical_cols])
```

```python
#performing train test split
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(x_encoded, y, test_size=0.2 , random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
#evaluation of the model
from sklearn.metrics import mean_absolute_error, mean_squared_error , r2_score
import numpy as np
```

```python
#Model 2 - Random forest
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100 , random_state=42)
rf_model.fit(X_train , y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
```

```python
mae_rf = mean_absolute_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest model Results:")
print("MAE:", mae_rf)
print("RMsE:", rmse_rf)
print("R2 score:", r2_rf)
```

### Conclusion
The inclusion of a randomly assigned city feature resulted in a slight decrease in model performance.
This indicates that location information must be based on real-world salary distributions to be useful.
The experiment successfully validated the modelâ€™s ability to handle additional categorical features.
